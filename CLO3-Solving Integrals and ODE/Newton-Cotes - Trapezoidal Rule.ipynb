```python
%matplotlib notebook
import matplotlib
from matplotlib.widgets import Slider
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set(color_codes=True)
sns.set_style("darkgrid")
```

```python
x = np.linspace(-0.01,1,100)
def equationf(x):
    y = 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5
    return y

#boundaries
a = 0
b = 0.8
n = 5 # number of trapeziums we want to do
ab = np.linspace(a,b,1000)
h = (b-a)/n
```

Below is a shwocase of the trapezium approximation method with a single application, the read area is the trapizium shape form which shall be taken as the integral of the function, while hte blue area is the area under the curve and both of which are boudned by a and b which are the regions we want, notice the overlap or lack thereof in bad approximations
