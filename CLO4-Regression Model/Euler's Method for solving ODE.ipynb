```python
%matplotlib inline
import matplotlib
import seaborn as sns
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from sympy import *

sns.set(style="darkgrid")
```

```python
x = np.linspace(0,4,100)
def equationf(x):
    ans = -2*x**3 + 12*x**2 -20*x + 8.5
    return ans
def equationftrue(x):
    y = -0.5*x**4 + 4*x**3 -10*x**2 + 8.5*x + 1
    return y
#boundaries
h = 0.5

xinitial = 0
xfinal  = 4

# conditions
startval = 1
initialcond = 1
 # initial val
```

```python
divalues = int((xfinal - xinitial)/h)
def linspaceval(h):
    divalues = int((xfinal - xinitial)/h)
    test = np.linspace(xinitial, xfinal , divalues +1)
    return test

def approx(h):
    i = 0
    val = 0
    yval = [initialcond]
    while i < (xfinal - xinitial)/h:
        ycalc = yval[i] + equationf(val)*h
        yval.append(ycalc)
        i += 1
        val +=h
    return yval

# test = diff(equationf(x), x)
# test.subs(x,5) 
```
